```python
from particle_propagation import Particle_Propagator, Approximated_Propagator, Thor_SCSI_Propagator
import numpy as np
import matplotlib.pyplot as plt

propagator = Particle_Propagator(Thor_SCSI_Propagator)
```

```python
import tqdm

def create_interpolation_files(x_input,px_input,propagator):
    """
    create grid for interpolation
    """
    Z1  = np.zeros((len(x_input),len(px_input)))   #Z1, for new x value
    Z2  = np.zeros((len(x_input),len(px_input)))   #Z2, for new px value
    
    for i in tqdm.tqdm(range(len(x_input))):

        next_x, next_px = propagator.propagation_single_round(np.array(len(px_input)*[x_input[i]]), 
                                                              px_input, deterministic = True)
        Z1[i,:] = next_x
        Z2[i,:] = next_px
    
    return Z1, Z2
```

```python
x_list = np.linspace(-45e-3,24.5e-3,1280)
px_list = np.linspace(-0.003,0.003,1280)
Z1,Z2 = create_interpolation_files(x_list,px_list,propagator)
```

```python
X,Y = np.meshgrid(x_list,px_list)
plt.figure(figsize=(9,5))
plt.pcolormesh(X,Y,Z1.T)
plt.colorbar()
```

```python
idx = (np.abs(Z1)>1)+(np.abs(Z2)>1)
Z1[idx] = float("nan")
Z2[idx] = float("nan")
```

```python
X,Y = np.meshgrid(x_list,px_list)
plt.figure(figsize=(9,5))
plt.pcolormesh(X,Y,Z1.T)
plt.colorbar()
```
